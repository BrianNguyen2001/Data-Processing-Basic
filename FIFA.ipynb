```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_json(r'https://raw.githubusercontent.com/BrianNguyen2001/Data-Processing-Basic/main/1.%20Data/FIFA_WorldCup2018Men.json')
```

```python
date_time = data.match_date.to_list()
```

```python
df = data[['home_team','away_team','home_score','away_score','match_week','stadium','referee']]
df.index = data.match_date
```

```python
df.head()
```

```python
# Xử lý dữ liệu cross
home_name, group, home_manager_name, home_manager_country = [],[],[],[]
away_name, away_manager_name, away_manager_country = [],[],[]
referee_name, referee_country = [],[]

for i in range(len(df)):
    home_name.append(df.home_team[i]['home_team_name']) # Name home team
    group.append(df.home_team[i]['home_team_group']) # Group
    home_manager_name.append(df.home_team[i]['managers'][0]['name']) # managers' name
    home_manager_country.append(df.home_team[i]['managers'][0]['country']['name']) # managers' country
    referee_name.append(df.referee[i]['name']) # name referee
    referee_country.append(df.referee[i]['country']['name']) # country referee
    away_name.append(df.away_team[i]['away_team_name']) # Name home team
    away_manager_name.append(df.away_team[i]['managers'][0]['name']) # managers' name
    away_manager_country.append(df.away_team[i]['managers'][0]['country']['name']) # managers' country
```

```python
finish = pd.DataFrame({'Date': date_time,
                       'Week': df.match_week.to_list(),
                       'Group': group,
                       'Home team': home_name,
                       'Away team': away_name,
                       'Home score': df.home_score.to_list(),
                       'Away score': df.away_score.to_list(),
                       'Home Managers': home_manager_name,
                       'Away Managers': away_manager_name,
                       'Home Managers country': home_manager_country,
                       'Away Managers country': away_manager_country,
                       'Referee name': referee_name,
                       'Referee country': referee_country})
```

```python
# Sắp xếp lại dataframe theo week và bảng đấu
finish.sort_values(['Week','Group'],inplace=True)

# Reset index của dataframe
finish.reset_index(drop=True,inplace=True)
```

```python
process = finish.copy()
```

```python
diff = process['Home score'] - process['Away score']
```

```python
process.head()
```

```python
winner_list = []
for i in range(len(process)):
    if diff[i] > 0:
        winner = process['Home team'][i]
    elif diff[i] < 0: 
        winner = process['Away team'][i]
    else:
        winner = 'Draw'
    winner_list.append(winner)
```

```python
process.insert(7, 'Difference', diff)
process.insert(8, 'Winner', winner_list)
```
